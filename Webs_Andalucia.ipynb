```python

df['Materia'] = df['Materia'].map(categorias)
df
```

```python
import matplotlib.pyplot as plt

# Añadir algunas entradas con baja frecuencia

df = pd.DataFrame(df)

# Conteo de frecuencia para la columna 'Materia'
materia_counts = df['Materia'].value_counts()

# Seleccionar las categorías con menos de un umbral específico para agruparlas en 'Otros'
threshold = 25 # Este umbral se puede ajustar según sea necesario
others_count = materia_counts[materia_counts < threshold].sum()
materia_counts = materia_counts[materia_counts >= threshold]

# Agregar la categoría 'Otros' con el conteo acumulado de las categorías menos frecuentes
materia_counts['Otros'] = others_count

# Crear gráfico de barras horizontales con los datos actualizados
materia_counts.plot(kind='barh')

# Invertir el eje y para que la barra con el mayor conteo esté arriba
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Suponiendo que 'Estado' es una columna en tu DataFrame
estado_counts = df['Estado'].value_counts()
estado_counts.plot(kind='bar')
plt.title('Distribución por Estado')
plt.xlabel('Estado')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Suponiendo que 'Tamaño' es numérico
plt.hist(df['Tamaño'], bins=10)  # Ajusta el número de bins según sea necesario
plt.title('Distribución del Tamaño')
plt.xlabel('Tamaño')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculando el conteo de valores en la columna 'Materia'
materia_counts = df['Materia'].value_counts()

# Manteniendo solo los 6 primeros y agrupando el resto bajo 'Otros'
top_materias = materia_counts[:6]
otros = materia_counts[6:].sum()
top_materias['Otros'] = otros

# Creando el gráfico de pastel con porcentajes enteros
plt.figure(figsize=(8, 8))
plt.pie(top_materias, labels=top_materias.index, autopct='%1.0f%%', startangle=140)
plt.title('Distribución de las 10 Principales Materias con Umbral')
plt.show()
```
